## Importation des données

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("activities.csv")
pd.set_option('display.max_columns', None)
df.head(40)
```

## Nettoyage des données

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
# On enlève les colonnes avec une majorité de valeurs nulls
df=df.dropna(thresh=250,axis=1)
df.shape
```

```python
# extraction de l'heure
df["Heure"]=df["Date de l'activité"].str.extract(r"(\d\d:\d\d:\d\d)")
# extraction de la date
mois={"janv.":1,"févr.":2,"mars":3,"avr.":4,"mai":5,"juin":6,"juil.":7,"août":8,"sept.":9,"oct.":10,"nov.":11,"déc.":12}
df["Jour"]=df["Date de l'activité"].str.extract("^(\d{1,2})")
df["Mois"]=df["Date de l'activité"].str.extract(" ([a-z,â,é,û]{3,4}\.?) ")
df["Mois_num"]=df["Mois"].map(mois)
df["Annee"]=df["Date de l'activité"].str.extract("(\d{4})")
df["Date"]=pd.to_datetime(dict(year=df["Annee"], month=df["Mois_num"], day=df["Jour"]))
df["Day"]=df["Date"].dt.dayofweek
df.head()
```

```python
# Selection des activités de 2020
df2020=df[df["Annee"]=="2020"]
df2020.head()
```

```python
# Effaçage activité avec un temps trop long (bug) :
df2020=df2020[df2020["Temps écoulé"]<24*60*60]
# Conversion de la distance de string à float :
df2020["Distance"]=df2020["Distance"].replace("\D","",regex=True).astype(float)/100
# Conversion du temps en minute (plus parlant que des secondes)
df2020["Temps écoulé"]=df2020["Temps écoulé"]/60
```

## Exploration des données

#### Quel est le sport le plus pratiqué en nombre d'activitées ?

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=45)
sns.barplot(df2020["Type d'activité"].value_counts().index,
            df2020["Type d'activité"].value_counts(),
            ).set_title("Number of activities per type");
```

#### Quelle est la répatition distance/temps en fonction du type d'activité ?

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df2020,x="Temps écoulé",y="Distance",hue="Type d'activité")
plt.title("distance en fonction du temps par type d'activitéé")
plt.ylabel("Distance (km)")
plt.xlabel("Temps");
```

* Les sorties à vélo virtuel (vélo d'intérieur) sont le plus rapides mais pas les plus longues
* Les sorties à vélo en extérieur sont les plus longue en temps et distance
* La course à pied est plus lente que le vélo (surprenant !)
* La marche est moins rapide que la course (surprenant également !)
* Les activitées type "entrainement" n'ont pas distance car statiques

##### Quel est le mois le plus actif de l'année ?

```python
tupleMois=tuple(mois.keys())
fig=plt.figure(1,figsize=(10,10))
# nb of activity :
plt.subplot(2,2,1)
df2020.groupby(["Mois_num","Type d'activité"])["Distance"].count().unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.xlabel("")
plt.ylabel("Nombre d'activitées total")
plt.title("Distance par mois")
plt.xticks(np.arange(12), tupleMois)
plt.title("Nombre d'activitées par mois")
# total distance :
plt.subplot(2,2,2)
df2020.groupby(["Mois_num","Type d'activité"])["Distance"].sum().unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.ylabel("Distance cumulée (km)")
plt.xticks(np.arange(12), tupleMois)
plt.title("Distance par mois")
plt.xlabel("Mois")
# total time :
plt.subplot(2,2,3)
ax=df2020.groupby(["Mois_num","Type d'activité"])["Temps écoulé"].sum().unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.ylabel("Temps cumulé (mins)")
plt.title("Temps par mois")
plt.xticks(np.arange(12), tupleMois)
plt.xlabel("Mois")

plt.legend(loc = (1.5 ,0), ncol=1);

```

#### A quoi une semaine "moyenne" ressemble elle ?

```python
fig=plt.figure(1,figsize=(10,10))
joursSem=("Lundi","Mardi","Mercredi","Jeudi","Vendredi","Samedi","Dimanche")
# nb of activity :
plt.subplot(2,2,1)
df2020.groupby(["Day","Type d'activité"])["Distance"].count().apply(lambda x:x/52).unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.ylabel("Nombre d'activitées moyen")
plt.xlabel("")
plt.xticks([])
plt.title("Nombre d'activitées par jours de la semaine")
# total distance :
plt.subplot(2,2,2)
df2020.groupby(["Day","Type d'activité"])["Distance"].sum().apply(lambda x:x/52).unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.ylabel("Distance moyenne (km)")
plt.title("Distance par jour de la semaine")
plt.xticks(np.arange(7), joursSem)
plt.xlabel("Mois")
# total time :
plt.subplot(2,2,3)
df2020.groupby(["Day","Type d'activité"])["Temps écoulé"].sum().apply(lambda x:x/52).unstack().plot(kind="bar",stacked=True,ax=plt.gca(),legend=None)
plt.ylabel("Durée moyenne (mins)")
plt.title("Durée par jour de la semaine")
plt.xticks(np.arange(7), joursSem)
plt.title("Durée d'activitée par jour de la semaine")

plt.legend(loc = (1.5 ,0), ncol=1);
```
